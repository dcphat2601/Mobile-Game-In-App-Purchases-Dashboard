```
from google.colab import files
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
csv_path = '/content/drive/MyDrive/game/mobile_game_inapp_purchases.csv'
df = pd.read_csv(csv_path)
```

```
print(df.shape)
df.head()
```

```
df.info()
df.isna().sum()
```

```
df['LastPurchaseDate'] = pd.to_datetime(df['LastPurchaseDate'], errors='coerce')
```

```
# Đếm missing
missing = df.isna().sum().sort_values(ascending=False)
print("\nMissing values per column:")
print(missing[missing > 0])

```

```
# 1. Drop
df = df.dropna(subset=['InAppPurchaseAmount'])

# 2. Fill
df['PaymentMethod'] = df['PaymentMethod'].fillna('Unknown')

# 3. Fill
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Gender'] = df['Gender'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Device'] = df['Device'].fillna('Unknown')
df['GameGenre'] = df['GameGenre'].fillna('Unknown')

# 4. Kiểm tra lại
print(df.isna().sum())
```

```
df.to_csv('/content/clean_mobile_game_data.csv', index=False)
```
